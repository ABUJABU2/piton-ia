```python
import openai
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv
import base64
```

```python
_ = (load_dotenv())  # emcontramos a API no arquivo .env e carregamos ela no nosso ambiente
client = OpenAI()    #instanciando s classe OpenAI no objeto client
```

```python
#criando a função de codificar a imagem em números
def encode_image(imagem):
    with open (imagem, "rb") as img:  #comando para abrir o arquivo lido "rb" ---> read bytes
        return base64.b64encode(img.read()).decode('utf-8') # retorno final da função que pega o arquivo aberto e decodifica a 
    #imagem em número no formato utf- 
```

```python
caminho = "socrates.jpeg"
base_64_img = encode_image(caminho)
```

```python
texto = input("escreva a sua pergunta sobre a imagem: ")
```

```python
#configurando a resposta do meu modelo de ia, utilizando o metodo create
#definindo os argumentos model e messages (formato de dicionario,sendo o conteudo em lista de dict)
resposta = client.chat.completions.create(
    model='gpt-5.2',
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": texto},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{base_64_img}"
                    }
                }
            ]
        }
    ],
)          
```

```python
print(resposta)
```
